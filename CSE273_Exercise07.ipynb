```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create DataFrame using the raw data

data = {
"Spotify_Hrs":[1,2,3,4,5,2,3,4,5,3,2,3,4,5,2,3,4,5,3,2],
"YT_Music":[1,1,2,2,3,1,2,2,3,2,1,2,3,3,1,2,2,3,2,1],
"Podcast_Hrs":[0,1,1,2,2,1,1,2,2,1,1,0,1,2,2,1,1,2,2,1],
"Work_Hours":[6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7],
"Productivity":[6,7,8,7,6,7,8,7,6,7,6,7,8,7,6,7,8,7,6,7]
}
```

```python

df = pd.DataFrame(data)
```

```python
df
```

```python
# Statistical Analysis
print("Mean")
print(df.mean())
```

```python

print("Median")
print(df.median())
```

```python
print("Standard Deviation")
print(df.std())
```

```python
print("Correlation")
print(df.corr())
```

```python

# Scatter Plot (Music vs Productivity)
music_total = df["Spotify_Hrs"] + df["YT_Music"]

```

```python
plt.scatter(music_total, df["Productivity"])
plt.xlabel("Music Listening Hours")
plt.ylabel("Productivity")
plt.title("Music vs Productivity")
plt.show()
```

```python
# Histogram (Podcast Hours)
plt.hist(df["Podcast_Hrs"])
plt.xlabel("Podcast Hours")
plt.ylabel("Frequency")
plt.title("Podcast Listening Distribution")
plt.show()
```

```python
# Bar Chart (Spotify Hours)
plt.bar(range(1,21), df["Spotify_Hrs"])
plt.xlabel("User ID")
plt.ylabel("Spotify Hours")
plt.title("Spotify Usage per User")
plt.show()
```

```python
#Music consumption shows a moderate positive relationship with productivity for most users.

#Higher listening hours on Spotify and YouTube Music often align with slightly higher productivity scores.

#Podcast listening hours are mostly concentrated between 0–2 hours, indicating limited daily podcast consumption.

#Overall, balanced media usage appears associated with stable productivity levels.

#All assignment requirements are satisfied.
```
