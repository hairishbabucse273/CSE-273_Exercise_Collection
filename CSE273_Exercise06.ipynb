```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name' : ['Aman', 'Riya', 'Rahul', 'Sneha', 'Arjun', 'Priya', "Karan", 'Neha'],
    'Age': [22, 25, 24, 23, 26, 27, 28, 22],
    'Gender': ['Male', 'Female', 'Male','Female','Male','Female','Male','Female'],
    'Marks': [75, 88, 92, 70, 85, 90, 78, 95]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.plot(df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(df["Name"],df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(df["Age"],bins = 5)
plt.xlabel("Age Distrubtion")
plt.xlabel("Age")
plt.ylabel("Frequncey")
plt.show()
```

```python
plt.scatter(df["Age"],df["Marks"])
plt.title("Age verses Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()
gender_counts =df["Gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("Gender Distribution of Students")
plt.show()

```

```python
plt.bar(df["Name"],df["Marks"],color="lavender")
plt.title("Student MArks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age of Distribution")
plt.subplot(1,2,2)
plt.bar(df["Name"],df["Marks"])
plt.title("Marks by Student")
plt.show()
```
